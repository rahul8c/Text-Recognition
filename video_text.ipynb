```
if not os.path.exists('image_frames'):
      os.makedirs('image_frames')
```

```
tvi = cv2.VideoCapture('Testvid.mp4')

index = 0
while  tvi.isOpened():
      ret,frame = tvi.read()
      if not ret:
         break
      name = '/content/image_frames/frame'+ str(index) +'.jpg'

      print('Extracing frames....' + name)
      cv2.imwrite(name, frame)
      index = index + 1
      if cv2.waitKey(5) & 0xFF == ord('q'):
       break
       
tvi.release()
cv2.destroyAllWindows()


```
